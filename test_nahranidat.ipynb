```python
import ast

def contains_at_least_half(ingredients_series):
    # Convert string representation of list to actual list
    try:
        ingredients_list = ast.literal_eval(ingredients_series)  # Use ast.literal_eval for safety
    except (ValueError, SyntaxError):
        return False  # Return False if the format is incorrect
        # Convert to set for faster intersection
    ingredients_set = set(ingredients_list)
    # Calculate the number of matching ingredients
    matching_ingredients_count = len(ingredients_set & desired_ingredients)
    
    # Check if the count of matching ingredients is at least half of the total desired ingredients
    return matching_ingredients_count >= len(desired_ingredients) / 2
```

```python
import ast  
def contains_at_least_half_with_partial_match(ingredients_series):
    # Convert string representation of list to actual list
    try:
        ingredients_list = ast.literal_eval(ingredients_series)  # Use ast.literal_eval for safety
    except (ValueError, SyntaxError):
        return False  # Return False if the format is incorrect

    # Convert each ingredient in the recipe list to lowercase for case-insensitive matching
    ingredients_list = [ingredient.lower() for ingredient in ingredients_list]

    # Count the number of desired ingredients that are present in any form in the ingredients list
    matching_ingredients_count = sum(
        any(desired_ingredient in ingredient for ingredient in ingredients_list) 
        for desired_ingredient in desired_ingredients   
    )
    
    # Check if the count of matching ingredients is at least half of the total desired ingredients
    return matching_ingredients_count >= len(desired_ingredients) / 2

# Apply the function and filter rows
filtered_indices = df[df['NER'].apply(contains_at_least_half_with_partial_match)].index.tolist()

# Display the resulting indices
print(filtered_indices)
```

```python
from collections import Counter

ingredient_counter = Counter()

# Iterate over each row in the 'ingredients' column
for ingredients_series in df['NER']:
    # Parse the string representation of the list
    try:
        ingredients_list = ast.literal_eval(ingredients_series)
    except (ValueError, SyntaxError):
        continue  # Skip rows with malformed data

    # Convert ingredients to lowercase for uniformity
    ingredients_list = [ingredient.lower().strip() for ingredient in ingredients_list]

    # Update the counter with the ingredients in this recipe
    ingredient_counter.update(ingredients_list)

# Get the most common ingredients
most_common_ingredients = ingredient_counter.most_common(50)  # Adjust the number to get top N ingredients

# Display the most common ingredients and their counts
for ingredient, count in most_common_ingredients:
    print(f"{ingredient}: {count}")
```
